```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

data = pd.read_csv('../data/mean_channel_data_with_button.csv')
targets = data['Phenotype']
del data['Subject']

HCdata = data[data['Phenotype'] == 'HC']
ADdata = data[data['Phenotype'] == 'AD']

del data['Phenotype']

labels = list(data.columns.values)
```

```python
# distplots for every feature  TODO: display graphs in facet grid instead of long line

for label in labels:
    axis = plt.subplots()[1]
    sns.distplot(HCdata[label].dropna(), ax=axis, color='g')
    sns.distplot(ADdata[label].dropna(), ax=axis, color='r')
```

```python
# correlation matrix

correlations = data.corr()

ax = plt.subplots(figsize=(16, 16))[1]

mask = np.zeros_like(correlations, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(correlations, mask=mask, square=True, ax=ax)
```
