```python
import pandas as pd
from IPython.display import HTML
import numpy as np

data = pd.read_csv('../data/full_data.csv')
labels = list(data.columns.values)
del labels[0]
del labels[0]
target = data['Phenotype']
del data['Subject']
del data['Phenotype']
```

```python
from sklearn import preprocessing
from sklearn import feature_selection

# sklearn doesn't have t-test, so looking at anova f-values. scipy has t-test (TODO?)

data = preprocessing.Imputer().fit_transform(data, target)
f_values = feature_selection.f_classif(data, target)[0]
HTML(pd.DataFrame(data=f_values, index=labels).sort_values(0, ascending=False).to_html())
```
